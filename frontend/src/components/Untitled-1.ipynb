```python
# Open EOPF Sentinel-2 L2A Zarr dataset
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# Zarr URL for the product
zarr_url = "https://objects.eodc.eu:443/e05ab01a9d56408d82ac32d69a5aae2a:202603-s02msil2a-eu/19/products/cpm_v262/S2A_MSIL2A_20260319T104041_N0512_R008_T31TDG_20260319T185813.zarr"

# Open the dataset
print("Opening Sentinel-2 L2A dataset...")
ds = xr.open_datatree(zarr_url, engine="zarr")
print(ds)

# Access individual bands at 10m resolution
print("\nLoading RGB bands (10m resolution)...")
red = xr.open_dataset(f"{zarr_url}/measurements/reflectance/r10m/b04", engine="zarr")
green = xr.open_dataset(f"{zarr_url}/measurements/reflectance/r10m/b03", engine="zarr")
blue = xr.open_dataset(f"{zarr_url}/measurements/reflectance/r10m/b02", engine="zarr")

# Extract data arrays
red_data = red.values.astype(float)
green_data = green.values.astype(float)
blue_data = blue.values.astype(float)

# Create true color composite (normalize to 0-1)
rgb = np.stack([red_data, green_data, blue_data], axis=-1)
rgb_normalized = rgb / 10000.0  # Sentinel-2 data is in 10000x scale
rgb_normalized = np.clip(rgb_normalized, 0, 1)

# Plot the true color image
plt.figure(figsize=(12, 10))
plt.imshow(rgb_normalized)
plt.title("Sentinel-2 L2A True Color Image - Barcelona (31TDG)")
plt.colorbar(label="Reflectance")
plt.axis('off')
plt.tight_layout()
plt.savefig("barcelona_s2_true_color.png", dpi=150, bbox_inches='tight')
plt.show()

print("\nImage saved as 'barcelona_s2_true_color.png'")
```
